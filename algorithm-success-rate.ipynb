# Titanic Machine Learning - Algorithm example
Titanic Algorithm success rate demonstration

```python

# Import libraries
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Import data

url = "https://raw.githubusercontent.com/pmarcelino/datasets/master/titanic-00.csv"
df = pd.read_csv(url)
df
```

```python
 # View the relationship between the class in which passengers were traveling and their age
 sns.boxplot(x='PClass', y='Age', data=df)
```

```python
# Define the independent and dependent variables
X = df[['PClass', 'Age', 'Sex']]
y = df['Survived']
```

```python
# Split the data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Define the algorithm of the model
model = RandomForestClassifier(random_state=9)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predicitions
y_pred = model.predict(X_test)
```

```python
# Evaluate model using success rate
accuracy_score(y_test, y_pred)
# The result demonstrates that the model has a success rate of approximately 80% (0.788). This means that, on average, if we know the class,
# the age and the sex of 100 different passengers, we are able to accurately predict what happened to 80 passengers.
# More explanation on: https://colab.research.google.com/drive/1LYbrENYHiAgZs-wFelxy9M8x_Ut17HWy
```
